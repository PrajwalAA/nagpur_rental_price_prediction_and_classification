```python
import requests
import speech_recognition as sr
import pyttsx3

# 🔑 Your OpenRouter API key
API_KEY = "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"  # replace with your real key
API_URL = "https://openrouter.ai/api/v1/chat/completions"

# --- Initialize recognizer and TTS ---
recognizer = sr.Recognizer()
tts = pyttsx3.init()

def speak(text):
    """Convert text to speech"""
    tts.say(text)
    tts.runAndWait()

def listen():
    """Capture voice and convert to text"""
    with sr.Microphone() as source:
        print("🎤 Listening...")
        audio = recognizer.listen(source)
    try:
        query = recognizer.recognize_google(audio)
        print(f"👉 You said: {query}")
        return query
    except sr.UnknownValueError:
        return "Sorry, I could not understand."
    except sr.RequestError:
        return "Network error."

def ask_api(query):
    """Send query to OpenRouter API"""
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "model": "openai/gpt-3.5-turbo",  # you can change to gpt-4 or other models
        "messages": [{"role": "user", "content": query}]
    }
    try:
        response = requests.post(API_URL, headers=headers, json=data)
        response.raise_for_status()
        result = response.json()
        return result["choices"][0]["message"]["content"]
    except Exception as e:
        return f"Error: {e}"

# --- Run chatbot loop ---
while True:
    query = listen()
    if query.lower() in ["exit", "quit", "stop", "bye"]:
        speak("Goodbye!")
        break
    response = ask_api(query)
    print("🤖 Bot:", response)
    speak(response)
```

```python
import requests
import folium
from IPython.display import display

def get_coordinates(location):
    url = "https://nominatim.openstreetmap.org/search"
    params = {"q": location, "format": "json"}
    headers = {"User-Agent": "JupyterApp/1.0"}
    response = requests.get(url, params=params, headers=headers)
    
    try:
        data = response.json()
    except ValueError:
        print("❌ Error: Could not decode JSON.")
        return None
    
    if data:
        lat, lon = float(data[0]["lat"]), float(data[0]["lon"])
        return lat, lon
    return None

# 🔹 Example Location
location = "Nagpur, India"
coords = get_coordinates(location)

if coords:
    lat, lon = coords
    print(f"📍 Location: {location}")
    print(f"   Latitude: {lat}")
    print(f"   Longitude: {lon}")

    # Create interactive map
    m = folium.Map(location=[lat, lon], zoom_start=12)
    folium.Marker([lat, lon], popup=location, tooltip="Click me!").add_to(m)

    # ✅ Force display in Jupyter
    display(m)
else:
    print("Location not found!")
```

```python
import pandas as pd
import requests
import folium
from IPython.display import display

# Function to get coordinates from Nominatim
def get_coordinates(location):
    url = "https://nominatim.openstreetmap.org/search"
    params = {"q": location, "format": "json"}
    headers = {"User-Agent": "JupyterApp/1.0"}
    response = requests.get(url, params=params, headers=headers)
    
    try:
        data = response.json()
    except ValueError:
        print(f"❌ Error decoding JSON for location: {location}")
        return None, None
    
    if data:
        lat, lon = float(data[0]["lat"]), float(data[0]["lon"])
        return lat, lon
    return None, None

# 🔹 Read Excel sheet
df = pd.read_excel("RD3_with_full_address.xlsx")  # Make sure 'full_address' column exists

# 🔹 Create new columns for latitude and longitude
latitudes = []
longitudes = []

for loc in df["full_address"]:
    lat, lon = get_coordinates(loc)
    latitudes.append(lat)
    longitudes.append(lon)

# Add new columns to the DataFrame
df["latitude"] = latitudes
df["longitude"] = longitudes

# Save updated Excel
df.to_excel("RD3_with_coordinates.xlsx", index=False)

# Display first few rows
df.head()

# 🔹 Optional: Display map for first address
if not df.empty and pd.notnull(df.loc[0, "latitude"]):
    m = folium.Map(location=[df.loc[0, "latitude"], df.loc[0, "longitude"]], zoom_start=12)
    for i, row in df.iterrows():
        if pd.notnull(row["latitude"]) and pd.notnull(row["longitude"]):
            folium.Marker(
                [row["latitude"], row["longitude"]],
                popup=row["full_address"],
                tooltip="Click me!"
            ).add_to(m)
    display(m)
```

```python
import pandas as pd
import requests
import folium
from IPython.display import display
import time  # to avoid hitting API too fast

# Function to get coordinates from Nominatim
def get_coordinates(location):
    url = "https://nominatim.openstreetmap.org/search"
    params = {"q": location, "format": "json"}
    headers = {"User-Agent": "JupyterApp/1.0"}
    response = requests.get(url, params=params, headers=headers)
    
    try:
        data = response.json()
    except ValueError:
        print(f"❌ Error decoding JSON for location: {location}")
        return None, None
    
    if data:
        lat, lon = float(data[0]["lat"]), float(data[0]["lon"])
        return lat, lon
    return None, None

# 🔹 Read Excel sheet with full_address column
df = pd.read_excel("RD3_with_full_address.xlsx")  # Make sure 'full_address' exists

# 🔹 Create new columns for latitude and longitude
latitudes = []
longitudes = []

for loc in df["full_address"]:
    lat, lon = get_coordinates(loc)
    latitudes.append(lat)
    longitudes.append(lon)
    time.sleep(1)  # 🔹 Wait 1 second to respect Nominatim usage policy

# Add new columns to DataFrame
df["latitude"] = latitudes
df["longitude"] = longitudes

# Save updated Excel
df.to_excel("RD3_with_coordinates.xlsx", index=False)

# 🔹 Display interactive map with all points
# Center map at the mean of all valid coordinates
valid_coords = df.dropna(subset=["latitude", "longitude"])
if not valid_coords.empty:
    map_center = [valid_coords["latitude"].mean(), valid_coords["longitude"].mean()]
    m = folium.Map(location=map_center, zoom_start=6)

    for _, row in valid_coords.iterrows():
        folium.Marker(
            [row["latitude"], row["longitude"]],
            popup=row["full_address"],
            tooltip="Click me!"
        ).add_to(m)

    display(m)
else:
    print("No valid coordinates to display on map.")
```
